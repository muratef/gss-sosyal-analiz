##### 2-) Veri Ön İşleme

```python

# 1. Dosya Yolları (VS Code dosya düzenine göre)
raw_path = '../data/raw/gss_2010_ham_full.parquet' # Parquet dosyanın bulunduğu yer
output_path = '../data/processed/temiz_veri.csv'

# 2. Belirlediğimiz 20 Kritik Sütun
hedef_sutunlar = [
    'year', 'realinc', 'income', 'educ', 'health', 'happy', 'class', 
    'prestg10', 'paeduc', 'age', 'sex', 'race', 'marital', 
    'polviews', 'trust', 'satfin', 'childs', 'wrkstat', 'region', 'natfare'
]

# 3. Veriyi Yükle ve Sütunları Seç
print("Veri yükleniyor...")
df = pd.read_parquet(raw_path)
df = df[hedef_sutunlar].copy()

# 4. GSS Spesifik Temizlik (IAP, DK, NA Değerlerini NaN Yapma)

# GSS'de genellikle 0, 8, 9 ve bunların katları geçersiz veridir.

print("Geçersiz değerler temizleniyor...")

df = df.replace([0, 8, 9, 98, 99], np.nan)

# 5. Temiz Veriyi Kaydet
df.to_csv(output_path, index=False)

print("-" * 30)
print(f"İşlem Tamamlandı!")
print(f"Kayıt Yeri: {output_path}")
print(f"Toplam Satır Sayısı: {len(df)}")
print(f"Sütun Sayısı: {len(df.columns)}")
```

Bu veri ön işleme aşaması, ham GSS veri setinden önceden belirlenen 20 kritik değişkeni ayırarak analizi odaklı hale getirir; veri setindeki "bilmiyorum" veya "cevap yok" anlamına gelen teknik dolgu değerlerini (0, 8, 9 vb.) standart boş değerlerle (NaN) değiştirerek modelin hatalı öğrenmesini engeller ve temizlenen veriyi bir sonraki aşama için CSV formatında dışa aktarır.

```python
dosya_yolu = '../data/processed/temiz_veri.csv'
# Dönüşüm sözlüğü (Tüm ihtimaller)
donusum = {2: 0, '2': 0, 'Kadın': 0, 'kadın': 0, 1: 1, '1': 1, 'Erkek': 1, 'erkek': 1}

# 1. Dosyayı ve df'yi güncelle
df = pd.read_csv(dosya_yolu)
df['sex'] = df['sex'].replace(donusum)
df['sex'] = pd.to_numeric(df['sex'], errors='coerce').fillna(0).astype(int)

# Güncellenmiş df'yi dosyaya kalıcı olarak yaz  
df.to_csv(dosya_yolu, index=False)
print("✅ temiz_veri.csv ve df güncellendi.")

# 2. df bellekte varsa onu da güncelle
if 'df' in locals():
    df.loc[:, 'sex'] = df['sex'].replace(donusum)
    df.loc[:, 'sex'] = pd.to_numeric(df['sex'], errors='coerce').fillna(0).astype(int)
    print("✅ df güncellendi.")
else:
    print("⚠️ df henüz oluşturulmamış, sadece df ve dosya güncellendi.")

# Kontrol
print("\nGüncel df Cinsiyet Dağılımı:")
print(df['sex'].value_counts())
```

Bu kod bloğu, sex sütunundaki kategorik veya karmaşık formatlı cinsiyet verilerini makine öğrenmesi modelinin işleyebileceği sayısal formatlara (0 ve 1) dönüştürür; yapılan bu güncellemeyi hem aktif çalışma belleğindeki veri çerçevesine hem de ana CSV dosyasına kalıcı olarak kaydederek veri setinde tutarlı bir sayısal standart sağlar.

```python
# Baskılanacak sayısal sütunları seçiyoruz
numeric_cols = ['realinc', 'age', 'educ', 'prestg10', 'childs']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Alt sınırın altındakileri alt sınıra, üst sınırın üstündekileri üst sınıra eşitle
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)

print("Aykırı değerler başarıyla baskılandı.")
```

Bu kod bloğu, belirlenen sayısal sütunlardaki aykırı değerleri (outliers) Çeyrekler Arası Aralık (IQR) yöntemini kullanarak tespit eder ve bu aşırı uç değerleri veri setinden silmek yerine belirlenen alt ve üst sınırlara baskılayarak (clipping) modelin bu uç sapmalardan olumsuz etkilenmesini engeller.
